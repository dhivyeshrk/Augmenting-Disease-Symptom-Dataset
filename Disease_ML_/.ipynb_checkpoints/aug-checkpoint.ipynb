```python

def nsum_all(nums, target_sum=0):
    def backtrack(index, current_sum, current_subset, target_sum):
        min_percentile_threshold = 40
        max_percentile_threshold = 60
        if target_sum == 0:
            percentile_low = np.percentile(nums, min_percentile_threshold)
            percentile_high = np.percentile(nums, max_percentile_threshold)
            elements_between_percentiles = [x for x in nums if x >= percentile_low and x <= percentile_high]
            target_sum = sum(elements_between_percentiles)
        
        if current_sum >= target_sum:
            result.append(current_subset[:])

        if index >= len(nums):
            return

        # Include the current number in the subset
        current_subset.append(nums[index])
        backtrack(index + 1, current_sum + nums[index], current_subset, target_sum)
        current_subset.pop()  # Backtrack and exclude the current number

        # Skip the current number and move to the next one
        backtrack(index + 1, current_sum, current_subset, target_sum)

    result = []
    backtrack(0, 0, [], target_sum)
    return result
```

```python

def fix_duplicate(data):
    data2 = data.copy()
    seen = set()
    for i in data2:
        while(i[0] in seen):
            i[0]+=0.1
        else:
            seen.add(i[0])
    return data2
def get_hashable_representation(lst):
    return tuple(sorted(lst))

def find_unique_combinations(data):
    unique_representations = set()
    unique_lists = []

    for sublist in data:
        hashable_representation = get_hashable_representation(sublist)
        if hashable_representation not in unique_representations:
            unique_representations.add(hashable_representation)
            unique_lists.append(sublist)

    return unique_lists
```

```python
import random

def random_subset_30_percent(main_list, seed):
    random.seed(seed)  # Set the random seed
    num_sublists_to_select = int(0.3 * len(main_list))
    shuffled_list = main_list.copy()  # Create a copy to avoid modifying the original list
    random.shuffle(shuffled_list)
    
    selected_sublists = shuffled_list[:num_sublists_to_select]
    
    return selected_sublists
```
